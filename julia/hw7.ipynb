```julia
a = ("Leo", 20, "2000/01/01", "#0000FF")

function swapping_first_and_last(tup::Tuple)
	a = collect(tup)
	a[1], a[end] = a[end], a[1]
	return Tuple(a)
end

swapping_first_and_last(a)
```

```julia
using Plots

poem = "桃花塢里桃花庵，桃花庵下桃花仙。 桃花仙人種桃樹，又摘桃花換酒錢。
酒醒只在花前坐，酒醉還來花下眠。 半醉半醒日復日，花落花開年復年。
但愿老死花酒間，不愿鞠躬車馬前。 車塵馬足顯者事，酒盞花枝隱士緣。
若將顯者比隱士，一在平地一在天。 若將花酒比車馬，彼何碌碌我何閑。
世人笑我太瘋癲，我笑他人看不穿。 不見五陵豪杰墓，無花無酒鋤作田。"

words = Set(poem)
counter = Dict(i => 0 for i in words)

for i in poem
	counter[i]+=1
end

sorted = sort([i for i in pairs(counter)], by=x->x.second, rev=true)
x, y = [i[1] for i in sorted], [i[2] for i in sorted]

println(length(counter))
println(sorted[1])
```

```julia
custom_font = "C:\\Windows\\Fonts\\KAIU.TTF"
Plots.font(custom_font, 12)
Plots.bar(1:5:5*84+1, y, xtickfont=font(custom_font, 5), bar_width=5, xticks=(1:5:5*84+1, x), size=(1000,200))
```
