```julia
using Pkg
Pkg.add("PyPlot")
using PyPlot
using IJulia
```

```julia
# const parameter
START, GOAL = 0, 1000
RABBIT_REST, RABBIT_RESTART = (GOAL-START)*.7, (GOAL-START)*.9
```

```julia
# define the rolling function
turtle_roll = () -> rand(1:1:6) # lambda
rabbit_roll = () -> rand(2:2:12)
```

```julia
# rolling test
turtle_roll(), rabbit_roll()
```

```julia
# to the start line
turtle, rabbit = START, START
turtle_history = [turtle]
rabbit_history = [rabbit]
twos_ = [2]
ones_ = [1]

# game update test
update_game = (msg, x_label) -> (
	IJulia.clear_output(true);
	title(msg);
	plot([GOAL GOAL GOAL GOAL],[0 1 2 3],"k--")
	plot(turtle_history,twos_,"g");
	plot(rabbit_history,ones_,"r");
	xlim(0, GOAL+10);
	ylim(0,3);
	xlabel(x_label);
	display(gcf());
)

while (turtle < GOAL) & (rabbit < GOAL)
	# roll dice
	turtle += turtle_roll()
	rabbit_resting = ((turtle>=RABBIT_RESTART) | (rabbit<RABBIT_REST))
	if rabbit_resting rabbit += rabbit_roll() end

	# update the plot
	turtle_history = append!(turtle_history, [turtle])
	twos_ = append!(twos_, [2])
	rabbit_history = append!(rabbit_history, [rabbit])
	ones_ = append!(ones_, [1])
	update_game("who will win?", "Turtle: " * string(turtle) * ", Rabbit: " * string(rabbit) * (if rabbit_resting "" else "(rest)" end))

	sleep(0.001)
end
update_game(((if (turtle>=GOAL) "turtle" else "rabbit" end) * " win!"), "Turtle: " * string(turtle) * ", Rabbit: " * string(rabbit))
IJulia.clear_output(true);
display(gcf());
println(if (turtle>=GOAL) "turtle" else "rabbit" end, " win!")
```
