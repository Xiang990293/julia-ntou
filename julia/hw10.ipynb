```julia
A = readdlm("user\\sst_data.txt")
year = A[:,1]
month = A[:, 2]
sst = A[:,3]


g = matopen("sst_data.mat", "w")
write(g, "year", year)
write(g, "month",month)
write(g, "sst", sst)

plotsst = matopen("sst_data.mat")
year = read(plotsst, "year")
month = read(plotsst, "month")
sst = read(plotsst, "sst")
t = year + month./12
plot(t, sst)

close(g)
close(plotsst)
display(gcf())
```
